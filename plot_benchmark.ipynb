```python
import pandas as pd
import math
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('random_keys.csv')
def plot_data(data):
    data_by_iter_num = data.groupby('NUM_ITERATORS')
    for iter_num in data_by_iter_num:
        avg_items = iter_num[1]['NUM_ITEMS'].mean()
        avg_comp = iter_num[1]['COMP_COUNT'].mean()
        avg_l_comp = iter_num[1]['LEARNED_COMP_COUNT'].mean()
        avg_error = iter_num[1]['CDF_ABS_ERROR'].mean()

        data_to_plot = iter_num[1][['LEARNED_COMP_COUNT', 'COMP_COUNT', 'CDF_ABS_ERROR']]
        data_to_plot.plot();
        plt.title("NUM_ITERATORS = " + str(iter_num[0]) + ", NUM_ITEMS = " + str(math.ceil(avg_items)))
        plt.show();
        
        def savings(v1, v2):
            return (v2 - v1) / v2 * 100.0

        print("Avg comp :                " + str(math.ceil(avg_comp)))
        print("Avg learned comp          " + str(math.ceil(avg_l_comp)))
        print("Avg error                 " + str(math.ceil(avg_error)))
        print("Avg learned comp + error: " + str(math.ceil(avg_l_comp + avg_error)) + "  %: " + str(savings(avg_l_comp + avg_error, avg_comp)))
```

```python
data = pd.read_csv('random_keys.csv')
plot_data(data)
```

```python
data = pd.read_csv('zipf.csv')
plot_data(data)
```
